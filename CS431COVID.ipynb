```python
df = pd.read_csv('us_covid_daily_reports.csv')
df["Coordinates"] = list(zip(df["Lat"], df["Long_"]))
df.head()
df.info()
```

```python
def analyze_df(df, head_rows=5):
    print(f"1. First {head_rows} rows of data:\n")
    display(df.head())

    print(f"\n 2. Data Info:\n")
    display(df.info())

    print(f"\n 3. Description of Data:\n")
    display(df.describe().transpose())

    print(f" 4. Missing Values: \n")
    missing_values = df.isnull().sum()
    display(missing_values)
```

```python
analyze_df(df, head_rows = 5)
```

```python
# Percentage of missing values
print(df.isnull().sum() / len(df) * 100)
```

```python
#cleaning lat and long null values since not too many
df_cleaned = df.dropna(subset=['Lat', 'Long_', 'Last_Update'])
```

```python
print(df_cleaned.isnull().sum() / len(df) * 100)
```

```python
#drop columns with >70% missing values
df_cleaned = df_cleaned.drop(columns=["People_Hospitalized", "Hospitalization_Rate", "People_Tested", "Mortality_Rate", "Recovered", "Active"])
```

```python
print(df_cleaned.isnull().sum() / len(df) * 100)
```

```python
#convert data nd last_update to datetime objects for time-series analysis
df_cleaned["Date"] = pd.to_datetime(df_cleaned["Date"])
df_cleaned["Last_Update"] = pd.to_datetime(df_cleaned["Last_Update"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = df_cleaned

df.hist(figsize=(20, 8), bins=30)
plt.tight_layout()
plt.show()

#distributions of the different numerical values to look for outliers
```

```python
#correlation of different features

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])  # Ensure it's a datetime object
df_grouped = df.groupby("Date")[["Confirmed", "Deaths"]].sum()

plt.figure(figsize=(12, 6))
plt.plot(df_grouped.index, df_grouped["Confirmed"], label="Confirmed Cases")
plt.plot(df_grouped.index, df_grouped["Deaths"], label="Deaths", linestyle="dashed")
plt.legend()
plt.title("Confirmed Cases and Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
top_countries = df.groupby("Province_State")["Confirmed"].sum().nlargest(20)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.xticks(rotation=45)
plt.title("Top 10 States by Confirmed Cases")
plt.show()
```

```python
top_countries = df.groupby("Province_State")["Deaths"].sum().nlargest(20)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.xticks(rotation=45)
plt.title("Top 10 States by Deaths")
plt.show()
```

```python
top_countries = df.groupby("Province_State")["Case_Fatality_Ratio"].sum().nlargest(20)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.xticks(rotation=45)
plt.title("Top 10 States by Case_Fatality_Ratio")
plt.show()
```
